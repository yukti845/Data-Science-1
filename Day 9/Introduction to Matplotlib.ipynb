## Matplotlib Example Workflow

```python
# 0 . import matplotlib and get it ready for plotting in jupyter

%matplotlib inline 
import matplotlib.pyplot as plt

# 1. Prepare data 
x = [1, 2, 3, 4]
y = [11, 22, 33, 44] 

# 2 Setup Plot 
fig, ax = plt.subplots(figsize=[10, 10])

# 3. Plot Data 
ax.plot(x, y)

# 4. Customize Plot 
ax.set (title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & Show ( You Save the whole figure )

fig.savefig("./images/sample-plot.png")
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt

x=[1, 3, 5, 7, 11, 13]
y=[2, 4, 6, 8, 10, 12]
fig, ax = plt.subplots(figsize=[5, 10])
ax.plot(x, y)
ax.set (title="Text Plot",
       xlabel="x-axis",
       ylabel="y-axis")
fig.savefig("./images/test_plot.png")
```

# Making figure with numpy arrays 

We Want :

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# plot the data and create a line plot

fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig , ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from dictionary 

nut_butter_prices = {"Almost butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig.ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store")
```
