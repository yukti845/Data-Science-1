```python
# Signature: plt.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
# Docstring:
# Plot y versus x as lines and/or markers.

plt.plot([1, 2, 3, 4])
```

```python
x = [1, 2, 3, 4]
y = [11, 12, 13, 14]
plt.plot(x, y);
```

```python
plt.plot()
```

```python
# 1st Method 

fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd Method 

fig = plt.figure()
ax = fig.add_axes([1, 1, 1, 1])
ax.plot(x, y)
plt.plot()
```

```python
# 3rd Method ( Recommended )

fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250]);
plt.show()
```
