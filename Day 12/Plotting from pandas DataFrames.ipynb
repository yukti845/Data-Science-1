```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv("heart_disease.csv")
heart_disease.head()
```

```python
# Create a histogram of age
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.head()
```

```python
heart_disease.plot.hist(subplots=True);
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50 ]
over_50.head()
```

```python
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target')
```

```python
# oo method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
```

```python
# OO method mixed with pyplot method 

fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim([45, 100])
```

```python
over_50.target.values
```

```python
# O method  from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
```

```python
# O method  from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

# Customize the plot 
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="cholesterol")
```

```python
# O method  from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
# Customize the plot 
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="cholesterol")

# Add a legend

ax.legend(*scatter.legend_elements(), title="Target");
```

```python
# O method  from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data 
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
# Customize the plot 
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="cholesterol")

# Add a legend

ax.legend(*scatter.legend_elements(), title="Target")

# Add ahorizontal line

ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```
